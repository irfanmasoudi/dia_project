### Click functions

that is a distribution function for the number of clicks given the bid for a specific customer class.
we decided to simulate it as follows:
$$click\_function(x|m) = m(1-e^{-4x+3x^{3}})$$
where 
* m is an upperbound for each category 

```python
advertising_env = Advertising_Config_Manager(adv_env_id)

n_arms = 10
min_bid = 0.0
max_bid = 1.0
feature_labels = advertising_env.feature_labels
bids = np.linspace(min_bid, max_bid, n_arms)
x = np.linspace(0,max(bids), num=550)
```

```python
plt.figure(figsize=(14,8))

for i,label in enumerate(feature_labels):
    y = advertising_env.click_functions[label](x)
    y_scatter = advertising_env.click_functions[label](bids)
    plt.plot(x, y, color=colors[i], label = label)
    plt.scatter(bids, y_scatter, color=colors[i])
    plt.title("Click Functions")
    plt.xlabel("Bid (â‚¬)")
    plt.ylabel("Number of Clicks")
    plt.legend()
```

### Conversion rates

The conversion rate of each class have been modelled in a matrix of probabilities located in the configuration file pricing_env.json.\
They have been chosen to best represent the willingness of the different types of users to purchase the product at the various price values.

```python
pricing_env = Pricing_Config_Manager(price_env_id)

prices = pricing_env.prices         # list of prices
categories = pricing_env.categories # list of categories

plt.figure(figsize=(14,8))
for i in range(len(categories)):
    y = pricing_env.probabilities[i]
    smooth = interp1d(prices, y, kind='cubic')
    plt.plot(prices, smooth(prices), color=colors[i], label = aux_dict[i])
    plt.scatter(prices, y, color=colors[i])
    plt.title("Conversion Rates")
    plt.xlabel("Price (â‚¬)")
    plt.ylabel("Conversion Rate")

plt.legend()
plt.show()
```

## Optimization Problem

### Assumptions

1. once a user makes a purchase with a price p, then the ecommerce will propose the same price p to future visits of the same user and this user will surely buy the item.
2. The revenue function must take into account the cost per click, while there is no budget constraint
3. the values of the parameters are known
4. the number of bid values are 10 as well as the number of price values

### Formal Model

*  $\max_{x_{j,t}, p_{j,t}} \sum_{j=1}^N v_{j}(x_{j,t}, p_{j,t}) n_{j}(x_{j,t})$
  
  such that:
  * $ v_{j} = ((p_{j,t} * d_{j}(p_{j,t})) (1 + \sum_{i=0}^9 i * Pr[ncb_{j,i}])) - c_{j}(x_{j,t})$ 
  
  where

  * $N$ number of subcampaigns
  * $C = \left\{C_{1}, ..., C_{N}\right\}$ advertising campaign
  * $T$ time horizon 
  * $t \in T$ instant of time 
  * $x_{j,t}$ bid of subcampaign *$C_{j}$* at time *t*
  * $p_{j,t}$ price of subcampaign *$C_{j}$* at time *t*
  * $v_{j}$ value per click (impression) of subcampaign *$C_{j}$*
  * $n_{j}(x_{j})$ number of clicks (impressions) of subcampaign   *$C_{j}$* given the values of bid $x_{j}$
  * $d_{j}(p_{j})$ conversion rate (the probability that a user will buy the item) of subcampaign   *$C_{j}$* given the price       $p_{j}$
  * $ncb_{j,i}$ the event, when customer of subcampaign   *$C_{j}$*, come back i times 
  * $Pr[ncb_{j,i}]$ a distribution probability over *$ncb_{j,i}$*
  * $c_{j}(x_{j})$ cost per click of subcampaign *$C_{j}$* given the values of bid $x_{j}$

### Algorithm for experience 1

In each instant of time $t \in T$:
1. The algorithm receives in input a table of values $v_{j}(x_{j,t}, p_{j,t}) n_{j}(x_{j,t})$ for each value of bid $x_{j,t}$ and price $p_{j}$
2. The optimal revenue can be found by looking for the maximum value in that input table

For the experiment 2 :

At first we have have estimates of the different parameters(number of clicks, cost per click, conversion rate and number of come backs of users) and with these estimates we compute the estimated reward for each price and bid and select the joints price and bid which have the best reward according to the estimated parameters. 

These bid and price are effective for one day, a number of clicks is chosen, a cost per click too, and for each click, an experience is conducted, whether the user who clicked buys or not and if he does, how many times he comes back. 

At the end of the day, we update all estimations of the 4 parameters thanks to the samples we had. And according to the new estimates we choose a new price and bid. 

This process is repeated for one year.

At the end we have estimated all parameters.
